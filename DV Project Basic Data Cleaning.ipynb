```python
clean_names = []

for hero_name in list(top_100_marvel['name']):
    if hero_name.strip() == 'Human Torch (Android)':
        clean_names.append(hero_name.strip())
        continue
    name_splits = hero_name.strip().split('(')
    
    cleaned_name = name_splits[0].strip()
    if cleaned_name not in clean_names:
        clean_names.append(cleaned_name)
    else:
        print (name_splits)
            
len(clean_names)
```

```python
top_100_marvel['name'] = clean_names
```

```python
top_100_marvel.head()
```

```python
writer = pd.ExcelWriter('./data/top_100_marvel_v2.xlsx')
top_100_marvel.to_excel(writer, 'Sheet1')

writer.save()
```
