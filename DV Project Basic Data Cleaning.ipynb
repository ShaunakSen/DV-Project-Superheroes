```python
superpowers_df = pd.read_csv('./data/super_hero_powers.csv')

superpowers_df.head()
```

```python
final_data = {'Name': [], 'Super Power': []}
```

```python
superpowers = list(superpowers_df.columns[1:])
```

```python
superpowers_df.iloc[0,0]
```

```python
for x in range(len(superpowers_df)):
    character = superpowers_df.iloc[x, 0].strip()
    for y in range(1,len(superpowers_df.columns)):
        super_power_present = superpowers_df.iloc[x, y]
        if super_power_present:
            final_data['Name'].append(character)
            final_data['Super Power'].append(superpowers[y-1])
        continue
            

```

```python
len(final_data['Super Power']) == len(final_data['Name'])
```

```python
superpowers_df_new = pd.DataFrame.from_dict(final_data)
```

```python
superpowers_df_new.head(20)
```

```python
set([len(power) for power in superpowers])
```

```python
too_long = []
for power in superpowers:
    if (len(power) > 22):
        if power not in too_long:
            too_long.append(power)
        
```

```python
too_long
```

```python
too_short = []
for power in superpowers:
    if (len(power) < 8):
        if power not in too_short:
            too_short.append(power)
print (too_short)
```

```python
too_short_mod = {}

for word in too_short:
    chars = list(word)
    mod_word = ""
    for char in chars:
        mod_word = mod_word + " " + char
    mod_word = mod_word.strip()
    too_short_mod[word] = mod_word
    
too_short_mod
```

```python
too_long_mod = {
    'Elemental Transmogrification' : 'Elem Transmogrification',
    'Probability Manipulation': 'Prob Manipulation',
    'Toxin and Disease Control': 'Toxin,Disease Control',
    'Toxin and Disease Resistance': 'Toxin,Disease Resis',
    'Mind Control Resistance': 'Mind Control Resis'
}
```

```python
superpowers_mod = []

for power in superpowers:
    if power in too_long_mod.keys():
        superpowers_mod.append(too_long_mod[power])
    elif power in too_short_mod.keys():
        superpowers_mod.append(too_short_mod[power])
    else:
        superpowers_mod.append(power)
```

```python
len(superpowers_mod)
```

```python
len(superpowers)
```

```python
superpowers_column_new = []

for power in list(superpowers_df_new['Super Power']):
    if power in superpowers_mod:
        superpowers_column_new.append(power)
    elif power in too_long_mod.keys():
        superpowers_column_new.append(too_long_mod[power])
    elif power in too_short_mod.keys():
        superpowers_column_new.append(too_short_mod[power])
    else:
        print (power)
```

```python
len(superpowers_column_new) == len(superpowers_df_new)
```

```python
superpowers_df_new['Super Power Processed'] = superpowers_column_new
```
